### **grouping by sub-journey in the trip**

```python
df["Segment_key"]=df["trip_uuid"]+','+df["source_center"]+'_'+df["destination_center"] 
segment_cols=['segment_actual_time', 'segment_osrm_time', 'segment_osrm_distance'] 
for col in segment_cols:
  df[col + '_sum']=df.groupby('Segment_key')[col].cumsum()
```

```python
creat_segment_dic={
    'data' : 'first',
    'trip_creation_time': 'first',
    'route_schedule_uuid': 'first', 
    'route_type' : 'first',
    'trip_uuid' : 'first',
    'source_center' : 'first',
    'source_name' : 'first', 
     
     'destination_center':'last',
     'destination_name':'last', 
     
     'od_start_time': 'first', 
     'od_end_time' : 'first',
     'start_scan_to_end_scan' : 'first', 
     
     'actual_distance_to_destination' :'last',
     'actual_time':'last', 
     
     'osrm_time' : 'last',
     'osrm_distance' : 'last',
     
      'segment_actual_time' : 'last', 
      'segment_osrm_time' : 'last', 
      'segment_osrm_distance':'last'
      }
```

### **Calculate the time taken between od_start_time and od_end_time and keep it as a feature. Drop the original columns, if required**

```python
# Calculatin the new feacher as od_final_time
segment['diff_h'] = (segment["od_start_time"].dt.hour - segment["od_end_time"].dt.hour)
segment['diff_m'] = (segment["od_start_time"].dt.minute - segment["od_end_time"].dt.minute)
segment['diff_m'] = round(segment['diff_m']/60, 2)
segment['od_final_time'] = segment['diff_h'].astype(float) + segment['diff_m']
```

### **Compare the difference between Point a. and start_scan_to_end_scan. Do hypothesis testing/ Visual analysis to check.**

```python
segment["od_final_time"].mean()
```

```python
segment["start_scan_to_end_scan"].mean()
```

```python
sns.scatterplot(segment["od_final_time"],segment["start_scan_to_end_scan"]) # Visual we can see od_final_time and start_scan_to_end_scan are not correlated
                                                                            # with each other
```

```python
#Let's do ttest
#H0: od_final_time and start_scan_to_end_scan are correlated with each other.
#H1: od_final_time and start_scan_to_end_scan are not correlated.
```

```python
from scipy.stats import ttest_ind
a=segment["od_final_time"]
b=segment["start_scan_to_end_scan"]
t_test_p_value = ttest_ind(a,b)[1]
print("t-test p-value:", t_test_p_value)
```

```python
# alpha is 0.05 significant value
if t_test_p_value < 0.05:
    print("Reject the null hypothesis H0")
    print("Hence Sastically proven that od_final_time and start_scan_to_end_scan are not correlated.")
else:
  print("fail to reject the null hypothesis H0")
  print("Hence Sastically proven that od_final_time and start_scan_to_end_scan are correlated.")
```

## **Hypothesis testing/ visual analysis between actual_time aggregated value and OSRM time aggregated value**

```python
segment["actual_time"].mean()
```

```python
segment["osrm_time"].mean()
```

```python
## actual_time and osrm_time both are numerical data hence, let's do t-test and levene test
# H0: actual_time and osrm_time are not depend on each other.
# H1: actual_time and osrm_time are depend each other.
```

```python
from scipy.stats import levene, ttest_ind

# Aggregated Actual_time data
actual = segment["actual_time"]

# Aggregated orsm (open-source routing engine) time data
orsm = segment["osrm_time"]

# dropping the missing values of both the groups
actual = actual.dropna()
orsm   = orsm.dropna()

print("Number of actual datapoints", len(actual))
print("Variance of actual:",round(np.var(actual),2))
print("Number of orsm datapoints", len(orsm))
print("Variance of orsm:",round(np.var(orsm),2))

#levene test
levene_p_value = levene(actual, orsm, center='median')[1]
print("levene test p-value", levene_p_value)

t_test_p_value = ttest_ind(actual, orsm)[1]
print("t-test p-value:", t_test_p_value)
```

```python
# alpha is 0.05 significant value
if t_test_p_value < 0.05:
    print("Reject the null hypothesis H0")
    print("Hence Sastically proven that Actual_time and osrm_time both are depend on each other.")
else:
  print("fail to reject the null hypothesis H0")
  print("Hence Sastically proven that Actual_time and osrm_time both are not depend on each other.")
```

```python
# Visualisation represntation of above observation 
sns.scatterplot(segment["actual_time"],segment["osrm_time"]) # from graph we can say that actual_time and orst_time are positively correlated with each other
```

## **Handle the outliers using the IQR method**

```python
Q1=segment.actual_distance_to_destination.quantile(0.25)
Q3=segment.actual_distance_to_destination.quantile(0.75)
IQR=Q3-Q1
print(IQR)
lower_limit=Q1 - 1.5*IQR
Upper_limit=Q3 + 1.5*IQR 
print(lower_limit,Upper_limit)
```

```python
## Let's remove the outlier from start_scan_to_end_scan column
segment=segment[(segment.actual_distance_to_destination>lower_limit) & (segment.actual_distance_to_destination<Upper_limit)]
```

```python
columns_having_outlier=['start_scan_to_end_scan','actual_distance_to_destination',	'actual_time',	'osrm_time',	
                       'osrm_distance',	'segment_actual_time',	'segment_osrm_time',	'segment_osrm_distance']
for col in columns_having_outlier:
  Q1=segment[col].quantile(0.25)
  Q3=segment[col].quantile(0.75)
  IQR=Q3-Q1
  print(IQR)
  lower_limit=Q1 - 1.5*IQR
  Upper_limit=Q3 + 1.5*IQR 
  print(lower_limit,Upper_limit)
  segment=segment[(segment[col]>lower_limit) & (segment[col]<Upper_limit)]
```
